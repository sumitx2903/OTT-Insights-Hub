```python
import pandas as pd

# Load the data
from pathlib import Path

orig_path = Path('Data_Extraction_and_Cleaning/Extraction/Python_file_and_Extracted_CSV/movies_tmdb_api_raw.csv')
if orig_path.exists():
	df = pd.read_csv(orig_path)
else:
	# try to find the file anywhere under the current working directory
	matches = list(Path('.').rglob('movies_tmdb_api_raw.csv'))
	if matches:
		df = pd.read_csv(matches[0])
		print(f"Loaded CSV from {matches[0]}")
	else:
		raise FileNotFoundError(
			"Could not find 'movies_tmdb_api_raw.csv'. "
			"Please place the file in the repository or update the path in the notebook."
		)

# Drop duplicate rows
df = df.drop_duplicates()

# Drop rows with missing values in important columns (e.g., 'title', 'release_date')
df = df.dropna(subset=['title', 'release_date'])

# Convert 'release_date' to datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Fill missing values in less important columns with a placeholder or appropriate value
df = df.fillna({'genre': 'Unknown', 'budget': 0, 'revenue': 0})

# Remove rows with invalid dates if any
df = df[df['release_date'].notnull()]

# Reset index after cleaning
df = df.reset_index(drop=True)

# Display cleaned dataframe
print(df.head())
```
